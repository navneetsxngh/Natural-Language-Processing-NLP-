## **STEMMING**

In Grammer, inflection is the modification of a word to express different grammitical categories such as tense, case, voice, aspect, person, number, gender, and mood.

Stemming is a process of reducing inflection in words to their roor forms such as mapping a group of words to the same stem even if the stem itself is not a valid word in their Language.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = "walk walks walking walked"
stem_words(sample)
```

```python
text = "probably my alltime favorite movie a story of selflessness sacrifice and dedication to a noble cause but its not preachy or boring it just never gets old despite my having seen it some 15 or more times in the last 25 years paul lukas performance brings tears to my eyes and bette davis in one of her very few truly sympathetic roles is a delight the kids are as grandma says more like dressedup midgets than children but that only makes them more fun to watch and the mothers slow awakening to whats happening in the world and under her own roof is believable and startling if i had a dozen thumbs theyd all be up for this movie"
print(text)
```

```python
stem_words(text)
```

## **Lemmatization**

Lemmatization, unlike Stemming reduces the inflected words properly ensuring that the root word belongs to the language.

In Lemmatization root word is called Lemma. A lemma(plural lemmas or lemmata) is the canonical form, dictionary form, or citation form of a set of words.

```python
import nltk
from nltk.stem import WordNetLemmatizer

## Download required resources (run once)
# nltk.download('wordnet')
# nltk.download('punkt_tab')

wordnet_lemmatizer = WordNetLemmatizer()

sentence = "He was running and eating at same time. He has bad habit of swimming after playing long hours in the Sun."
punctuations = "?:!.,;"

# Tokenize and remove punctuations
sentence_words = nltk.word_tokenize(sentence)
for word in sentence_words:
    if word in punctuations:
        sentence_words.remove(word)

print(sentence_words)
print()
# Print Word vs Lemma
print("{0:20}{1:20}".format("Word", "Lemma"))
for word in sentence_words:
    print("{0:20}{1:20}".format(word, wordnet_lemmatizer.lemmatize(word, pos='v')))
```
