```python
## Preprocessing and Cleaning

# Postitve Review --> 1 and Negative review ---> 0
df['rating'] = df['rating'].apply(lambda x : 0 if x < 3 else 1)
```

```python
## Lower all the Cases
df['reviewText'] = df['reviewText'].str.lower()
```

```python
df.head()
```

```python
import re

def remove_special_characters(text):
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return text

df['reviewText'] = df['reviewText'].apply(remove_special_characters)
```

```python
df.sample(10)
```

```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
## Removing Stopwords
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    return " ".join(filtered_words)

df['reviewText'] = df['reviewText'].apply(remove_stopwords)
```

```python
df.sample(10)
```

```python
## Removing URL's

def remove_urls(text):
    url_pattern = r'https?://\S+|www\.\S+'
    return re.sub(url_pattern, '', text)

df['reviewText'] = df['reviewText'].apply(remove_urls)
```

```python
df.sample(10)
```

```python
## Removing HTML Tags
from bs4 import BeautifulSoup

def remove_html_tags(text):
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text()

df['reviewText'] = df['reviewText'].apply(remove_html_tags)
```

```python
df.sample(10)
```

```python
## Remove Additional Whitespaces
import re
def remove_extra_whitespace(text):
    text = re.sub(r'\s+', ' ', text)   
    return text.strip()                

df['reviewText'] = df['reviewText'].apply(remove_extra_whitespace)
```

```python
df.sample(10)
```

```python
## Lemmatizer
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
df['reviewText']=df['reviewText'].apply(lambda x:lemmatize_words(x))
```

```python
df.head()
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df['reviewText'],df['rating'],
                                              test_size=0.20)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow = CountVectorizer()
X_train_bow = bow.fit_transform(X_train).toarray()
X_test_bow = bow.transform(X_test).toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
X_train_tfidf = tfidf.fit_transform(X_train).toarray()
X_test_tfidf = tfidf.transform(X_test).toarray()
```

```python
from sklearn.naive_bayes import GaussianNB
nb_model_bow = GaussianNB().fit(X_train_bow,y_train)
nb_model_tfidf = GaussianNB().fit(X_train_tfidf,y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
y_pred_bow = nb_model_bow.predict(X_test_bow)
```

```python
y_pred_tfidf = nb_model_bow.predict(X_test_tfidf)
```

```python
confusion_matrix(y_test,y_pred_bow)
```

```python
print("BOW accuracy: ",accuracy_score(y_test,y_pred_bow))
```

```python
confusion_matrix(y_test,y_pred_tfidf)
```

```python
print("TFIDF accuracy: ",accuracy_score(y_test,y_pred_tfidf))
```
