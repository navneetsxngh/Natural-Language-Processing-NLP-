**Feature Extraction** from Text is a process in **Natural Language Processing (NLP)** where we convert text data (words, sentences, documents) into numerical features so that machine learning algorithms can understand and process it.

### **1. What is Feature Extraction from Text?**

Feature extraction means transforming text into numerical vectors that represent important information in the text.

Example:

Text:
`"I love machine learning"`

Possible numerical representation:
`[1, 1, 1, 0, 0]`

Each number represents whether certain words exist or how important they are.

Common techniques include:

- Bag of Words (BoW)
- TF-IDF
- Word Embeddings (Word2Vec, GloVe)
- Sentence Embeddings
- N-grams

### **2. Why Do We Need Feature Extraction?**

Machine learning models such as:

- Logistic Regression
- SVM
- Naive Bayes
- Neural Networks

cannot understand raw text.

Example ML input must look like this:

| Feature1 | Feature2 | Feature3 |
| -------- | -------- | -------- |
| 0.2      | 0.5      | 0.8      |

But text looks like: `"This movie is amazing"`

So we must convert text → numbers.

Feature extraction helps to:

1. Convert text into numeric form
2. Capture important words or patterns
3. Reduce noise
4. Improve model performance

Example use cases:

1. Spam detection
2. Sentiment analysis
3. Chatbots
4. Search engines
5. Document classification

### **3. Why is Feature Extraction Difficult?**

Text data is unstructured and ambiguous, which makes it challenging.

1. High Dimensionality

Language has thousands to millions of words.

Example vocabulary: `V = 50,000 words`

Feature vector size becomes 50,000 dimensions.

2. Synonyms

Different words can mean the same thing.

Example:

```
good
great
excellent
```
A simple algorithm might treat them as completely different.

3. Context Problem

Example:
```
"I like the movie"
"I do not like the movie"
```
The word "not" changes everything, but simple models may miss it.

4. Polysemy (Same word, different meaning)

Example:
```
bank = river bank
bank = financial bank
```
Meaning depends on context.

5. Word Order Matters

Example:
```
Dog bites man
Man bites dog
```
Bag-of-Words treats both exactly the same.

````python
import pandas as pd
````

## Bag of Words

````python
df = pd.DataFrame({
    'text' : ['people watch campusx',
              'campusx watch campusx',
              'people write comment',
              'campusx write comment'],
    'output' : [1,1,0,0]
})
df
````

````python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
````

````python
bow = cv.fit_transform(df['text'])
````

````python
print(cv.vocabulary_)
````

````python
print(bow[0].toarray())
print(bow[1].toarray())
````

````python
cv.transform(["campusx watch and write comment of campusx"]).toarray()
````
