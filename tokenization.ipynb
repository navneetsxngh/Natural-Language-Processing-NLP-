## 1. Use the `split()` Function

```python
## Word Tokenization
sent1 = "I am going to Delhi"
sent1.split()
```

```python
## Sentence Tokenization
sent2 = "I am going to Delhi. I will stay there for 3 days. Let\'s hope the trip would be great"
sent2.split('.')
```

```python
sent4 = "Where do you think I should go? I have 3 day holiday"
sent4.split('.')
```

## 2. Regular Expression

```python
import re
sent3 = "I am going to Delhi!"
tokens = re.findall(r"[\w']+", sent3)
tokens
```

```python
text = """Lorem Ipsum is simply dummy text of the printing and typesetting industry?
Lorem Ipsum has been the industry's standard dummy text ever since the 1500s,
when an unknown printer took a galley of type and scrambled it to make a type specimen book."""

# Split the text into sentences using regex
sentences = re.compile(r'[.!?] ').split(text)

print(sentences)
```

## 3. NLTK

```python
import nltk

# For newer NLTK versions (3.8.2+ / 3.9+)
nltk.download('punkt_tab')   # try this first

# For older versions, or if the above doesn’t work
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sent1 = "I am going to Delhi"
word_tokenize(sent1)
```

```python
text = """Lorem Ipsum is simply dummy text of the printing and typesetting industry?
Lorem Ipsum has been the industry's standard dummy text ever since the 1500s,
when an unknown printer took a galley of type and scrambled it to make a type specimen book."""

sent_tokenize(text)
```

## More Examples

```python
corpus = '''Hello Welcome, to Navneet Singh's GitHub Repository of learning NLP tutorials.
Please check out all the files in this repo! to become expert in NLP
'''
print(corpus)
```

```python
## Sentence --> Paragraphs
from nltk.tokenize import sent_tokenize
documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
  print(sentence)
```

```python
## Paragraph --> Words
## Sentence --> Words
from nltk.tokenize import word_tokenize
tokens = word_tokenize(corpus)
```

```python
tokens
```

```python
for sentence in documents:
  print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
